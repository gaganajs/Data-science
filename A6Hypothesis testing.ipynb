```python
# Hypothesis Testing - Weekly Operating Cost

import math
from scipy.stats import norm

# Given values
x_bar = 3050        # Sample mean
mu = 4000           # Theoretical mean
sigma = 125         # Standard deviation
n = 25              # Sample size
alpha = 0.05        # Significance level

# Step 1: Calculate Z statistic
z = (x_bar - mu) / (sigma / math.sqrt(n))

print("Z-statistic:", z)

# Step 2: Critical value (Right-tailed test)
z_critical = norm.ppf(1 - alpha)

print("Critical Z-value:", z_critical)

# Step 3: P-value
p_value = 1 - norm.cdf(z)

print("P-value:", p_value)

# Step 4: Decision
if z > z_critical:
    print("Reject Null Hypothesis (H0)")
else:
    print("Fail to Reject Null Hypothesis (H0)")

# Step 5: Conclusion
print("\nConclusion:")
print("There is no strong evidence to support that operating costs are higher than the model.")
```

```python
# Confidence Interval using t-distribution

import numpy as np
from scipy.stats import t

# Data
data = [1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32,
        0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29]

# Step 1: Basic calculations
n = len(data)
mean = np.mean(data)
std_dev = np.std(data, ddof=1)  # sample std

print("Sample Mean:", mean)
print("Sample Std Dev:", std_dev)

# Step 2: t critical value (99% CI)
alpha = 0.01
t_critical = t.ppf(1 - alpha/2, df=n-1)

print("t Critical Value:", t_critical)

# Step 3: Margin of error
margin = t_critical * (std_dev / np.sqrt(n))

# Step 4: Confidence Interval
lower = mean - margin
upper = mean + margin

print("\n99% Confidence Interval (using t-distribution):")
print("Lower Limit:", lower)
print("Upper Limit:", upper)
```

```python
# Confidence Interval using Z-distribution

from scipy.stats import norm

# Given population std
sigma = 0.2

# Step 1: Z critical value (99%)
z_critical = norm.ppf(1 - 0.01/2)

print("Z Critical Value:", z_critical)

# Step 2: Margin of error
margin = z_critical * (sigma / np.sqrt(n))

# Step 3: Confidence Interval
lower = mean - margin
upper = mean + margin

print("\n99% Confidence Interval (using Z-distribution):")
print("Lower Limit:", lower)
print("Upper Limit:", upper)
```
