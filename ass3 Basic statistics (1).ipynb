```python
# Step 1: Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Load the dataset
# Replace with your file path
df = pd.read_csv("C:\\Users\\gagan\\Downloads\\sales_data_with_discounts.csv")

# Step 3: Display first 5 rows
print("Dataset Preview:")
print(df.head())

# Step 4: Identify numerical columns
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
print("\nNumerical Columns:", num_cols)

# Step 5: Basic Statistics
print("\nMean:\n", df[num_cols].mean())
print("\nMedian:\n", df[num_cols].median())
print("\nMode:\n", df[num_cols].mode().iloc[0])
print("\nStandard Deviation:\n", df[num_cols].std())

# -------------------------------
# DATA VISUALIZATION
# -------------------------------

# Step 6: Histograms
for col in num_cols:
    plt.figure()
    sns.histplot(df[col], kde=True)
    plt.title(f"Histogram of {col}")
    plt.show()

# Step 7: Boxplots
for col in num_cols:
    plt.figure()
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot of {col}")
    plt.show()

# Step 8: Categorical Columns
cat_cols = df.select_dtypes(include=['object']).columns
print("\nCategorical Columns:", cat_cols)

# Step 9: Bar Charts for categorical columns
for col in cat_cols:
    plt.figure()
    df[col].value_counts().plot(kind='bar')
    plt.title(f"Bar Chart of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```
