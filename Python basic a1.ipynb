```python
# Simple program to check if a number is prime

num = int(input("Enter a number: "))

count = 0

for i in range(1, num + 1):
    if num % i == 0:
        count = count + 1

if count == 2:
    print("The number is Prime")
else:
    print("The number is Not Prime")
```

```python
# Program to check the product of two random numbers

import random

num1 = random.randint(1, 10)
num2 = random.randint(1, 10)

print("First number:", num1)
print("Second number:", num2)

answer = int(input("Enter the product of these numbers: "))

if answer == num1 * num2:
    print("Correct answer!")
else:
    print("Wrong answer.")
    print("The correct product is:", num1 * num2)
```

```python
# Hospital Billing System

room_charge = 2000 * 4
doctor_fee = 1500
lab_test = 300 * 3
medicine = 2400

total = room_charge + doctor_fee + lab_test + medicine

if total > 10000:
    discount = total * 0.10
else:
    discount = 0

final_bill = total - discount

print("Total Bill:", total)
print("Discount:", discount)
print("Final Bill:", final_bill)
```
