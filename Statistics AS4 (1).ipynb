```python
import pandas as pd
```

```python
patient_df = pd.read_csv("C:\\Users\\gagan\\Downloads\\Patient_Data.csv")
billing_df = pd.read_csv("C:\\Users\\gagan\\Downloads\\Billing_Data.csv")

print(patient_df.head())
```

```python
patient_df.info()
```

```python
patient_df = patient_df[['PatientID', 'Department', 'Doctor', 'BillAmount']]
print(patient_df.head())
```

```python
mean_value = patient_df['BillAmount'].mean()
patient_df['BillAmount'] = patient_df['BillAmount'].fillna(mean_value)
print(patient_df.head())
```

```python
patient_df = patient_df.drop_duplicates(subset='PatientID')
print(patient_df)
```

```python
dept_bill = patient_df.groupby('Department')['BillAmount'].sum()
print(dept_bill)
```

```python
merged_df = pd.merge(patient_df, billing_df, on='PatientID', how='left')
print(merged_df.head())
```

```python
new_patients = pd.DataFrame({
    'PatientID': [201, 202],
    'Department': ['Cardiology', 'Neurology'],
    'Doctor': ['Dr. Smith', 'Dr. John'],
    'BillAmount': [5000, 7000]
})

final_df = pd.concat([merged_df, new_patients], axis=0, ignore_index=True)

print(final_df.tail())
```

```python
billing_extra = pd.DataFrame({
    'InsuranceCovered': [True] * len(final_df),
    'FinalAmount': final_df['BillAmount'] * 0.9
})

final_df = pd.concat([final_df, billing_extra], axis=1)

print(final_df.head())
```
